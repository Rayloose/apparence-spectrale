```python
import numpy as np
from PIL import Image, ImageShow
import matplotlib.pyplot as plt
import pandas as pd
```

```python
images = ['IMG_3927.JPG', 'IMG_3932_decharge.jpg', 'IMG_3934_led2700K.jpg', 'IMG_3936_4000K.jpg', 'IMG_3938_6500K.jpg']

#liste des coordonnées en largeur et hauteur des centres approximatifs de chaque carré
list_x = [1240,2150, 3070, 3960, 4840, 5780]
list_y = [860, 1770, 2700, 3680]

order = [0, 4, 8, 12, 16, 20, 1, 5, 9 ,13, 17, 21, 2, 6, 10, 14, 18, 22, 3, 7, 11, 15, 19, 23]

def get_pixel(liste, x, y):
  return [liste[x + y*width][0], liste[x + y*width][1], liste[x + y*width][2]]


def moyenne_pixel(liste, x,y,m):
    '''(x, y) les coordonnées du pixel
       m le nombre de pixel qu'on prend de chaque côté du pixel principal pour faire la moyenne'''
    RGB = []
    for k in range (-m, m+1):
      for l in range (-m, m+1):
        RGB.append(liste[y+k][x+l])

    mean = np.mean(RGB, axis=0)

    return mean.astype(int) #tronquer les valeurs


def tab_colors(list_x, list_y, m=0):
  list_rgb = []
  for k in list_x :
    for j in list_y :
      list_rgb.append(moyenne_pixel(k,j,m))
      #if j==1600 and i==2150:
          #print(list_rgb)
      #list_rgb.append(get_pixel(penis,i,j))

  ordered_colors = [list_rgb[k] for k in order]

  colors = np.array_split([tuple(k) for i in ordered_colors], 4)
  plt.imshow(colors)
  #plt.title("éclairé sous " + 'ampoule à décharge')
  #print(colors)
  return ordered_colors

def show_tab(image):
  im = Image.open(image, 'r')

  width, height = im.size
  pixel_values = list(im.getdata())

  #transforme la liste de 1d en tableau de la taille de l'image. Chaque sous liste correspond à une ligne
  #de l'image avec à chaque une liste de 3 éléments [r, g, b]
  weirdpixel_values = np.array(pixel_values).reshape((height, width, 3))

  orderco = tab_colors(list_x,list_y,20)
  plt.savefig('éclairage ' + image, dpi = 600)
  dfcolors = pd.DataFrame(orderco, columns=list('rgb'))
  return dfcolors

for i in images :
  show_tab(i)
  #plt.savefig(i, dpi = 600)
```
